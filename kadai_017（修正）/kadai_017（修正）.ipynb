```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      return f"{self.name}（{self.age}歳）は大人です"

    else:
      return f"{self.name}（{self.age}歳）は大人ではありません"

humans = [
    Human("侍太郎", 36),
    Human("侍二郎", 30),
    Human("侍三郎", 19),
    Human("侍四郎", 16)
]

for human in humans:
  print(human.check_adult())
```
