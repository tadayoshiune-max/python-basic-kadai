```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      return f"{self.name}（{self.age}歳）は大人です"

    else:
      return f"{self.name}（{self.age}歳）は大人ではありません"

human1 = Human("侍太郎", 36)
human2 = Human("侍二郎", 30)
human3 = Human("侍三郎", 19)
human4 = Human("侍四郎", 16)

print(human1.check_adult())
print(human2.check_adult())
print(human3.check_adult())
print(human4.check_adult())
```
