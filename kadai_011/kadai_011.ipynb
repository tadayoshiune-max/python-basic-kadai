```python
planets = ["水", "金", "地", "火", "木", "土", "天", "海", "冥"]

# --- forループで出力 ---
print("【forループ】")
for p in planets:
    print(p)

# --- whileループで出力 ---
print("【whileループ】")

i = 0  # 最初の要素は0番
while True:
    if i < 9:           # 要素数は9個
        print(planets[i])
        i += 1          # 次の番号へ進む
    else:               # iが9以上になったら終了
        break
```
