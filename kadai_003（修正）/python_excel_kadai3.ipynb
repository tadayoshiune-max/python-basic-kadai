```python
import pandas as pd

df1 = pd.read_excel('2022_年間売上表.xlsx')
df2 = pd.read_excel('2023_年間売上表.xlsx')

df = pd.concat([df1,df2])

result = df.groupby(['商品', '売上年'])['金額（千円）'].sum().reset_index()

write = pd.ExcelWriter('売上集計表.xlsx')

result.to_excel(write, sheet_name = 'Sheet1', index = 0)
write.close()
```

```python
import openpyxl
from openpyxl.styles import PatternFill

wb = openpyxl.load_workbook('売上集計表.xlsx')
ws = wb.active

fill_pattern = PatternFill(patternType='solid', fgColor='F2F2F2')

for row in ws['A1:C1']:
    for cell in row:
        cell.fill = fill_pattern

wb.save('売上集計表.xlsx')
```
